# Installing WFA Cardinality Estiamtion Framework on Google CoLab

To install the framework clone the repository and run the installation.

Then restart the runtime. For that go to **Runtime** menu and select **Restart runtime** option.

You should be able to import the framework modules after that.

```
!git clone https://github.com/world-federation-of-advertisers/cardinality_estimation_evaluation_framework
!cd cardinality_estimation_evaluation_framework; pip install -r requirements.txt
!cd cardinality_estimation_evaluation_framework; python3 setup.py install
```

```
from wfa_cardinality_estimation_evaluation_framework.estimators import liquid_legions

sketch = liquid_legions.LiquidLegions(10, 50000, 42)

for i in range(123456):
  sketch.add_id(i)

print('Estimated cardinality:', sketch.get_cardinality())
```
